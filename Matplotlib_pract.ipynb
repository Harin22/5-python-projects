```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(10, 5), dpi=100)
plt.plot(x, y, color='red', linestyle='--', linewidth=2, marker='o', markersize=5, label='Sine Function')
plt.fill_between(x, y, color='lightblue', alpha=0.3)

plt.title('Sine Wave', fontsize=16, color='blue')
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)

plt.grid(True, color='gray', linestyle='-', linewidth=0.5)
plt.legend(loc='upper right')
plt.xlim(0, 10)
plt.ylim(-1.5, 1.5)

plt.xticks(np.arange(0, 11, 1))
plt.yticks(np.linspace(-1, 1, 5))

plt.annotate('Peak', xy=(np.pi/2, 1), xytext=(2, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.savefig('sine_wave.png', dpi=300)
plt.show()
```
